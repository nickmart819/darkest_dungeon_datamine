```python
import re
from collections import defaultdict
import pprint
import json
import pandas as pd
import os
pp = pprint.PrettyPrinter(depth=4)
# filepath = "D:\SteamLibrary\steamapps\common\DarkestDungeon\heroes\houndmaster\houndmaster.info.darkest"
```

```python
def parse_file_name_for_id(filepath):
    filename_without_extension = os.path.splitext(os.path.basename(filepath))[0]
    #naming convention of files ends with .info.darkest
    filename_without_extensions_extension = filename_without_extension.split(".")[0]
    return filename_without_extensions_extension
    # Expected output: houndmaster as an example
```

```python

def parse_info_file(filepath):
    parsed_data = defaultdict(list)
    with open(filepath) as file:
        for line in file:
            if line.strip() == "":
                continue
            
            #splits the out key word and then uses an index to pull that first word for our line entry type 
            entry_type = line.split(":")[0].strip()
            
            # pulls out the fields and their values
            field_pairs = re.findall(r"\.(\w+)\s+([\"].+?[\"]|\S+)", line)

            # Build a dictionary of field names and cleaned values
            entry_data = {
                field_name: field_value.strip('"')
                for field_name, field_value in field_pairs
            }

            parsed_data[entry_type].append(entry_data)

    parsed_data_json = json.dumps(parsed_data, indent=4, sort_keys=True)
    data = json.loads(parsed_data_json)

    

    return data


#TO DO NEXT
    # FIND A WAY TO UTILIZE THE FILE NAME TO USE AS AN ID TO ADD INTO EACH DATAFRAME SO WE KNOW WHAT CHARACTER IS ASSOCIATED WITH EACH ROW
    # CREATE AN ID ALONGSIDE THE FILE NAME TO CREATE A DIMENSION WITH
    # CREATE FUNCTION TO LOOP THROUGH EACH FILE DIRECTORY
```

```python
def add_to_df(data, filename_without_extensions_extension):
    df_armour_per_file = pd.DataFrame(data["armour"])
    df_armour_per_file.insert(0,'character',filename_without_extensions_extension)

    df_combat_move_skill_per_file = pd.DataFrame(data["combat_move_skill"])
    df_combat_move_skill_per_file.insert(0,'character',filename_without_extensions_extension)

    df_combat_skill_per_file = pd.DataFrame(data["combat_skill"])
    df_combat_skill_per_file.insert(0,'character',filename_without_extensions_extension)

    df_controlled_per_file = pd.DataFrame(data["controlled"])
    df_controlled_per_file.insert(0,'character',filename_without_extensions_extension)

    df_crit_per_file = pd.DataFrame(data["crit"])
    df_crit_per_file.insert(0,'character',filename_without_extensions_extension)

    df_deaths_door_per_file = pd.DataFrame(data["deaths_door"])
    df_deaths_door_per_file.insert(0,'character',filename_without_extensions_extension)

    df_generation_per_file = pd.DataFrame(data["generation"])
    df_generation_per_file.insert(0,'character',filename_without_extensions_extension)

    # df_id_index_per_file = pd.DataFrame(data["id_index"])
    # df_id_index_per_file.insert(0,'character',filename_without_extensions_extension)

    df_resistances_per_file = pd.DataFrame(data["resistances"])
    df_resistances_per_file.insert(0,'character',filename_without_extensions_extension)

    df_skill_selection_per_file = pd.DataFrame(data["skill_selection"])
    df_skill_selection_per_file.insert(0,'character',filename_without_extensions_extension)

    df_tags_per_file = pd.DataFrame(data["tag"])
    df_tags_per_file.insert(0,'character',filename_without_extensions_extension)

    df_weapon_per_file = pd.DataFrame(data["weapon"])
    df_weapon_per_file.insert(0,'character',filename_without_extensions_extension)

    return df_armour_per_file, df_combat_move_skill_per_file, df_combat_skill_per_file, df_controlled_per_file, df_crit_per_file, df_deaths_door_per_file, df_generation_per_file, df_resistances_per_file, df_skill_selection_per_file, df_tags_per_file, df_weapon_per_file

```

```python
# Final DataFrames as arguments
def add_data_to_final_dataframes(data, filename_without_extensions_extension, 
                                 final_armour_df, final_combat_move_skill_df, final_combat_skill_df,
                                 final_controlled_df, final_crit_df, final_deaths_door_df,
                                 final_generation_df, final_resistances_df, final_skill_selection_df,
                                 final_tags_df, final_weapon_df):
    # Get the dataframes for the current file
    dfs = add_to_df(data, filename_without_extensions_extension)

    # Append each dataframe to the corresponding final dataframe
    final_armour_df = pd.concat([final_armour_df, dfs[0]], ignore_index=True)
    final_combat_move_skill_df = pd.concat([final_combat_move_skill_df, dfs[1]], ignore_index=True)
    final_combat_skill_df = pd.concat([final_combat_skill_df, dfs[2]], ignore_index=True)
    final_controlled_df = pd.concat([final_controlled_df, dfs[3]], ignore_index=True)
    final_crit_df = pd.concat([final_crit_df, dfs[4]], ignore_index=True)
    final_deaths_door_df = pd.concat([final_deaths_door_df, dfs[5]], ignore_index=True)
    final_generation_df = pd.concat([final_generation_df, dfs[6]], ignore_index=True)
    final_resistances_df = pd.concat([final_resistances_df, dfs[7]], ignore_index=True)
    final_skill_selection_df = pd.concat([final_skill_selection_df, dfs[8]], ignore_index=True)
    final_tags_df = pd.concat([final_tags_df, dfs[9]], ignore_index=True)
    final_weapon_df = pd.concat([final_weapon_df, dfs[10]], ignore_index=True)

    # Return all the final dataframes
    return (
        final_armour_df, final_combat_move_skill_df, final_combat_skill_df,
        final_controlled_df, final_crit_df, final_deaths_door_df,
        final_generation_df, final_resistances_df, final_skill_selection_df,
        final_tags_df, final_weapon_df
    )
```

```python
def get_info_darkest_filepaths(base_directory):
    info_darkest_filepaths = []
    for root, dirs, files in os.walk(base_directory):
        for file in files:
            if file.endswith("info.darkest"):
                info_darkest_filepaths.append(os.path.join(root, file))
    return info_darkest_filepaths
```

```python
# Initialize final DataFrames outside the loop
final_armour_df = pd.DataFrame()
final_combat_move_skill_df = pd.DataFrame()
final_combat_skill_df = pd.DataFrame()
final_controlled_df = pd.DataFrame()
final_crit_df = pd.DataFrame()
final_deaths_door_df = pd.DataFrame()
final_generation_df = pd.DataFrame()
final_resistances_df = pd.DataFrame()
final_skill_selection_df = pd.DataFrame()
final_tags_df = pd.DataFrame()
final_weapon_df = pd.DataFrame()

# Loop through the files and update the final DataFrames
base_directory = "D:\\SteamLibrary\\steamapps\\common\\DarkestDungeon\\heroes\\"
filepaths = get_info_darkest_filepaths(base_directory)

for filepath in filepaths:
    filename_without_extensions_extension = parse_file_name_for_id(filepath)
    data = parse_info_file(filepath)
    
    #Have to initialize the final dataframes here so that they are not reset to empty each time
    final_armour_df, final_combat_move_skill_df, final_combat_skill_df, final_controlled_df, \
    final_crit_df, final_deaths_door_df, final_generation_df, final_resistances_df, \
    final_skill_selection_df, final_tags_df, final_weapon_df = add_data_to_final_dataframes(
        data, filename_without_extensions_extension,
        final_armour_df, final_combat_move_skill_df, final_combat_skill_df,
        final_controlled_df, final_crit_df, final_deaths_door_df,
        final_generation_df, final_resistances_df, final_skill_selection_df,
        final_tags_df, final_weapon_df
    )

# Print the final DataFrames
final_armour_df
```
